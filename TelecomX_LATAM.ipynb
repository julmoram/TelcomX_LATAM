```python
df_account.columns = df_account.columns.str.replace('PaperLessBilling', 'PaperlessBilling')
df_account.columns = df_account.columns.str.replace('Changes,', '')
df_internet = df_internet.replace('No internet service', 'No')
```

```python
df_account['Charges.Monthly'] = pd.to_numeric(df_account['Charges.Monthly'], errors='coerce')
df_account['Cuentas_Diarias'] = df_account['Charges.Monthly'] / 30


df_account[['Charges.Monthly', 'Cuentas_Diarias']].head()
```

```python
def binary_encode(column):
    return column.map({'Yes': 1, 'No': 0, 'No phone service': 0, 'No internet service': 0})

df_phone['PhoneService'] = binary_encode(df_phone['PhoneService'])
df_phone['MultipleLines'] = binary_encode(df_phone['MultipleLines'])

internet_columns = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                   'TechSupport', 'StreamingTV', 'StreamingMovies']
for col in internet_columns:
    df_internet[col] = binary_encode(df_internet[col])


datos['Churn'] = datos['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df_account.columns = df_account.columns.str.lower().str.replace('.', '_')
df_internet.columns = df_internet.columns.str.lower()
df_phone.columns = df_phone.columns.str.lower()


df_account = df_account.rename(columns={
    'charges_monthly': 'cargo_mensual',
    'charges_total': 'cargo_total',
    'cuentas_diarias': 'cargo_diario'
})
```

```python
contract_dummies = pd.get_dummies(df_account['contract'], prefix='contract')
payment_dummies = pd.get_dummies(df_account['paymentmethod'], prefix='payment')
internet_dummies = pd.get_dummies(df_internet['internetservice'], prefix='internet')

df_account = pd.concat([df_account, contract_dummies], axis=1)
df_account = pd.concat([df_account, payment_dummies], axis=1)
df_internet = pd.concat([df_internet, internet_dummies], axis=1)
```

```python
# Reemplazar espacios y valores vacÃ­os con NaN
df_account['cargo_mensual'] = df_account['cargo_mensual'].replace(r'^\s*$', np.nan, regex=True)
df_account['cargo_total'] = df_account['cargo_total'].replace(r'^\s*$', np.nan, regex=True)
df_account['cargo_diario'] = df_account['cargo_diario'].replace(r'^\s*$', np.nan, regex=True)

# Convertir a numÃ©rico
df_account['cargo_mensual'] = pd.to_numeric(df_account['cargo_mensual'], errors='coerce')
df_account['cargo_total'] = pd.to_numeric(df_account['cargo_total'], errors='coerce')
df_account['cargo_diario'] = pd.to_numeric(df_account['cargo_diario'], errors='coerce')

# Verificar valores nulos despuÃ©s de la conversiÃ³n
print(df_account[['cargo_mensual', 'cargo_total', 'cargo_diario']].isnull().sum())
```

**Informe Final: AnÃ¡lisis de EvasiÃ³n de Clientes (Churn) en TelecomX LATAM**  

**IntroducciÃ³n**
El **churn** (evasiÃ³n de clientes) es un problema crÃ­tico en telecomunicaciones porque afecta directamente los ingresos. Este anÃ¡lisis buscÃ³:  
- Identificar patrones de cancelaciÃ³n.  
- Descubrir quÃ© factores influyen en la decisiÃ³n de los clientes.  
- Proponer estrategias basadas en datos para reducir el abandono.  

**Dataset**: InformaciÃ³n de clientes de TelecomX LATAM (contratos, facturaciÃ³n, servicios y estado de churn).  



**impieza y PreparaciÃ³n de Datos**  
1. **OrganizaciÃ³n**:  
   - Los datos venÃ­an en JSON con columnas anidadas (`phone`, `internet`, `account`).  
   - Se normalizaron y unieron en un solo DataFrame (`df_completo`).  

2. **Limpieza**:  
   - Se corrigieron errores ("Tiber optic" â†’ "Fiber optic").  
   - Se eliminaron espacios en valores numÃ©ricos (`cargo_mensual`, `cargo_total`).  
   - Se convirtiÃ³ "Yes/No" a "1/0" para anÃ¡lisis matemÃ¡tico.  

3. **Transformaciones clave**:  
   - Nueva columna: `Cuentas_Diarias = cargo_mensual / 30`.  
   - EstandarizaciÃ³n de nombres (ej. `PaperLessBilling` â†’ `paperless_billing`).  



**Hallazgos Clave**  

**1. DistribuciÃ³n general de Churn**  
- **26% de los clientes cancelaron** el servicio.  
- **74% permanecieron**.  

**2. Factores con mayor impacto en la evasiÃ³n**:  
**Contratos mes a mes**: 45% de tasa de churn (vs. 8% en contratos de 2 aÃ±os).  
**Fibra Ã³ptica**: 30% de churn (vs. 18% en DSL).  
**Pagos con cheque electrÃ³nico**: 35% de churn (vs. 15% en pagos automÃ¡ticos).  

**3. RelaciÃ³n con gastos**:  
**Clientes que cancelan pagan mÃ¡s** en promedio ($75 vs. $60 mensuales).  
**Gasto total menor en clientes con churn**, lo que sugiere menor antigÃ¼edad.  

  

**Recomendaciones**  

**Acciones inmediatas**:  
 **Promover contratos largos**: Ofrecer descuentos en planes anuales.  
 **Incentivar pagos automÃ¡ticos**: 5% de bonificaciÃ³n.  
 **Mejorar la fibra Ã³ptica**: Revisar calidad del servicio y expectativas.  

**Acciones estratÃ©gicas**:  
 **Programa de fidelizaciÃ³n**: Beneficios para clientes antiguos.  
 **Sistema de alertas**: Monitorear clientes con:  
   - Contrato mes a mes + fibra Ã³ptica + gasto alto.  



**ConclusiÃ³n**  
Los clientes con contratos cortos, fibra Ã³ptica y pagos manuales son los que mÃ¡s cancelan. **Enfocarse en estos grupos** con estrategias personalizadas puede reducir la evasiÃ³n significativamente.
